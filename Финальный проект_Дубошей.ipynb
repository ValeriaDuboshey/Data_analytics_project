### Расчет трат клиентов по тарифам

Правила формирования тарифов:

Ежемесячная плата за тариф:

- Тариф А — 5000 ₽,
- B — 2000 ₽,
- С — 1000 ₽.

Входящие звонки — бесплатны на всех тарифах,

Исходящие звонки внутри сети — бесплатны до 2000 минут в месяц на всех тарифах. Цена сверх лимита:

- A — 0.1 ₽/мин.,
- B — 0.15 ₽/мин.,
- C — 0.3 ₽/мин.

Исходящие звонки, кроме внутренних:

- A — 0.4 ₽/мин.,
- B — 0.5 ₽/мин.,
- C — 0.7₽/мин.

За каждого оператора клиент платит 100 ₽ в месяц на любом тарифе — дополнительно к стоимости тарифа и цене звонков.

```python
#df_pr['month'] = pd.DatetimeIndex(df_pr['date']).month 
#df_pr['date']=pd.to_datetime(df_pr['date'])
df_pr['month']=df_pr['date'].astype('datetime64[M]')
df_pr
```

```python
df_grouped = df_pr.query('direction == "out"').groupby(['month', 
                                                        'user_id', 
                                                        'internal', 
                                                        'tariff_plan']).agg({'total_call_duration': 'sum'}).reset_index()
df_grouped_1 = df_pr.groupby(['month', 'user_id', 'tariff_plan']).agg({'operator_id': 'nunique'}).reset_index()
df_grouped = pd.merge (df_grouped, df_grouped_1, on=['month', 'user_id'], how='outer')
df_grouped['tariff_plan_x']=df_grouped['tariff_plan_x'].fillna(df_grouped['tariff_plan_y'])
df_grouped[['internal', 'total_call_duration']]=df_grouped[['internal', 'total_call_duration']].fillna(0)
df_grouped = df_grouped.drop('tariff_plan_y', 1)
df_grouped['total_call_duration'] = df_grouped['total_call_duration']/60
df_grouped['total_call_duration'] = df_grouped['total_call_duration'].apply(math.ceil)
df_grouped.rename({'operator_id':'count_operators', 'total_call_duration':'total_call_duration_min', 'tariff_plan_x':'tariff_plan'}, axis=1, inplace=True)
df_grouped
```

Мы сгруппировали данные по месяцам, коду клиента, тарифу и логу внутреннего/внешнего звонка. При этом в связи с тем, что входящие звонки бесплатны на всех тарифах - мы сначала создали таблицу только с операторами, которые совершают исходящие звонки, а потом добавили к ним клиентов, которые не попали в эту таблицу с операторами, которые только принимают звонки, для более корректного расчета выручки.

```python
tariffs=pd.DataFrame({"tariff_plan": ['A', 'B', 'C'], "fix_price": [5000, 2000, 1000], 
                      "int_extra_min_price": [0.1, 0.15, 0.3], "out_extra_min_price": [0.4, 0.5, 0.7], "operator_cost": [100, 100, 100]})
tariffs
```

```python
df_grouped = pd.merge(df_grouped, tariffs, on = 'tariff_plan', how='left')
df_grouped
```

```python
df_grouped['extra_costs_internal'] = np.where(df_grouped['internal'] == True, 
                                              (df_grouped['total_call_duration_min']-2000)*df_grouped['int_extra_min_price'], 0)
#Удалим отрицательные значения, которые получились в колонке 'extra_costs_internal
df_grouped.loc[(df_grouped['extra_costs_internal'] < 0), 'extra_costs_internal'] = 0
df_grouped['extra_costs_external'] = np.where(df_grouped['internal'] == False, 
                                              df_grouped['total_call_duration_min']*df_grouped['out_extra_min_price'], 0)
df_grouped
```

Проверим как посчитатись доп. расходы по внутренним звонкам

```python
#Посмотрим уникальные значения доп расходов
df_grouped['extra_costs_internal'].unique()
```

```python
#Доп расходов по внутренним разговоров не выявлено. 
#Проверим максимальное количество минут внутри сети, если менее 2000 - все правильно, доп расходов по внутренним звонкам нет.
df_grouped.query('internal == True')['total_call_duration_min'].max()
```

```python
df_grouped_with_internal = df_grouped
# Очистим данных от строк со звонками внутри сети
df_grouped=df_grouped.query('internal != True').reset_index(drop=True)
```

Рассчитаем итоговые расходы каждого клиента помесячно

```python
df_grouped['monthly_fee'] = df_grouped['fix_price']+df_grouped['count_operators']*df_grouped['operator_cost']+df_grouped['extra_costs_internal']+df_grouped['extra_costs_external']
df_grouped
```

### Выделение клиентов, которые пользуются неоптимальными тарифами и определение наилучшего тарифа


Для определения наиболее оптимального тарифа для клиентов пройдем следующие этапы:  
 1. Возьмем данные о длительности звонков активных клиентов в ноябре (берем последний месяц, тк данные предыдущих месяцев могут быть уже неактуальны для клиентов в связи с изменением кол-ва операторов или внутренних изменений компаний).  
 2. Рассчитаем расходы клиентов по каждому тарифу.  
 3. Выберем оптимальный тариф.  
 4. Сравним с текущим тарифом.  
 5. Определим клиентов, которые используют неоптимальный тариф.

1. Получим данные по активным клиентам

```python
active_clients = pd.merge(active_clients_list, df_grouped, on='user_id', how='left')
active_clients
```

Оставим только данные за ноябрь и уберем данные о внутренних звонках, тк как мы уже выяснили ни кто из клиентов не превысил лимит внутренних звонков.

```python
active_clients = active_clients.query('month == "2019-11-01" & internal!=True').reset_index(drop=True)
active_clients
```

 2. Рассчитаем расходы клиентов по каждому тарифу. 

```python
k=0
for tariff in tariffs['tariff_plan']:
    for i in range(len(active_clients)):
        if tariff != active_clients.loc[i,'tariff_plan']:
            active_clients.loc[i, 'costs_'+tariff] = (
                    tariffs.loc[k,'fix_price'] + 
                    tariffs.loc[k,'out_extra_min_price'] * 
                    active_clients.loc[i, 'total_call_duration_min'] +
                    active_clients.loc[i, 'count_operators'] *
                    active_clients.loc[i, 'operator_cost']
                )
            if active_clients.loc[i, 'costs_'+tariff] < active_clients.loc[i,'monthly_fee']:
                active_clients.loc[i, 'optimal_tariff'] = tariff
    k+=1
```

## Проверка гипотез

### В результате перехода ряда клиентов на более эффективный тариф вручка за ноябрь не изменится.

Нулевая гипотеза: выручка не изменится.  
Альтернативная гипотеза: выручка изменится.

```python
revenue_before=active_clients['monthly_fee'] 
revenue_after=active_clients['min_costs'] 
```

Так как мы хотим понять, равна ли выручка по тарифам, то за критический уровень статистической значимости примем 1%.  
Используем метод st.ttest_ind для подтверждения / опровержения гипотезы.

```python
alpha = 0.01

results = st.ttest_ind(revenue_before, revenue_after)
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print()
print('Средний чек до перехода клиентов: {:.2f}'.format(revenue_before.mean()))
print('Средний чек после перехода клиентов: {:.2f}'.format(revenue_after.mean()))
print()
print('Снижение среднего чека: {:.0%}'.format(1-revenue_after.mean()/revenue_before.mean()))
```

Как мы видим, средний чек в ноябре существенно снизится, если ряд клиентов перейдет на более выгодный тариф.

### Среднее количество минут внутри сети больше, чем среднее количество минут с другими операторами.

Возьмем датасет с активными клиентами за ноябрь и подготовим двет таблицы с продолжительностью звонков

```python
internal_duratial=active_clients_with_internal.query('internal == True')['total_call_duration_min']
external_duratial=active_clients_with_internal.query('internal != True')['total_call_duration_min']
mean_duration_internal = internal_duratial.mean()
```

Нулевая гипотеза: средняя продолжительность разговоров одинаковая.  
Альтернативная гипотеза: средняя продолжительность различается.  

Критический уровень статистической значимости примем 5%.   
Нас интересует различие продолжительности разговоров как в большую, так и в меньшую сторону, тогда используем метод st.ttest_1samp для подтверждения / опровержения двусторонней гипотезы.

```python
alpha = 0.05

results = st.ttest_1samp(external_duratial, mean_duration_internal)
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")  
    
    
print()
print('Средняя продолжительность внутри сети: {:.2f}'.format(mean_duration_internal))
print('Средняя продолжительность внешних звонков: {:.2f}'.format(external_duratial.mean()))
```
